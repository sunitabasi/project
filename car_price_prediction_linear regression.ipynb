```python
car_dataset = car_dataset.drop(['fuel_type'],axis = 1) #dropping original columns, axis =1 will add new column
car_dataset                                                                                # axis=0 will add new row
```

```python
#splliting the data and target
X = car_dataset.drop(['car_brand','city',"carprice['000']"],axis=1)
Y = car_dataset["carprice['000']"]
```

```python
#Splitting Training and Test data
```

```python
 from sklearn.model_selection import train_test_split
# # We specify this so that the train and test data set always have the same rows, respectively
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Model tranining- linear regression
# loading the linear regression model
from sklearn.linear_model import LinearRegression
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train,Y_train)

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X_train)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, Y_train)
```

```python
# prediction on Training data
from sklearn import metrics
training_data_prediction = lin_reg_model.predict(X_train)
# R squared Error-how well the model's predictions fit the actual data, with values closer to 1 indicating a better fit and values closer to 0 indicating a poorer fit.
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
#r square error indicates the goodness of fit of your regression model.
```

```python
training_data_prediction = pol_reg.predict(poly_reg.fit_transform(X_train))

error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
# it seems that there is high accuracy with linear model
```
